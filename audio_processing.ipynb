```python
import wave
import matplotlib.pyplot as plt
import numpy as np
```

```python
# audio format
# .mp3  (highly compressed)
#  flac  (low compressed)
#  .wave (best quality)


# audio signal parametes
# num of channels
# sample widht
# framerate
# num of frames
# values of frame

#  read .wave data
obj = wave.open("./10000.wav", 'rb')
# get the channel
print(f'the total channel is : {obj.getnchannels()}')
print(f'num of frames is {obj.getnframes()}')
print(f'the params are : {obj.getparams()}')

# get the duration
t_audio = obj.getnframes() / obj.getframerate()
# getframerate = num of framerate per second
print(f'the duration is {t_audio}')

#  get the total frames using readframe
#  devide by 2 because the .wav have 2 samples width
print(len(obj.readframes(-1)) / 2)
frames = obj.readframes(-1)
obj.close()
# make duplicte
# set the params
obj_new = wave.open("./10000_new.wav", "wb")
obj_new.setnchannels(1)
obj_new.setsampwidth(2)
obj_new.setframerate(360000,)
obj_new.writeframes(frames)
obj_new.close()
```

```python
# plotting 
wav = wave.open("./10000.wav", "rb")
freq = wav.getframerate()
n_freqs = wav.getnframes()
# 
signal_value = wav.readframes(-1)
wav.close()
t_audio = freq / n_freqs
# for y axes
signal_array = np.frombuffer(buffer = signal_value, dtype=np.int16)
#  for x axes
time = np.linspace(0, t_audio, num = n_freqs)
fig = plt.figure(figsize =(10, 6))
plt.plot(time, signal_array)
plt.title("Audio signal")
plt.xlabel('time (s)')
plt.ylabel("signal audio")
plt.xlim(0, t_audio)
# plt.ylim(0, np.max(signal_array))
plt.show()
```
